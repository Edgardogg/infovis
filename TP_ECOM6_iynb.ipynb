Analizando la cantidad de votos según su tipo (positivos, blancos, nulos..)

```python
df.groupby(["tipoVoto"]).agg({"votos":"sum"}).sort_values("votos", ascending=False).reset_index()
```

Analizando el tipo de voto por Cargo

Diputados Nacionales:

```python
df[df.Cargo == "DIPUTADOS NACIONALES"].groupby(["Cargo","tipoVoto"]).agg({"votos":"sum"}).sort_values(["Cargo","votos"], ascending=False)
```

Diputados Provinciales

```python
df[df.Cargo == "DIPUTADOS PROVINCIALES"].groupby(["Cargo","tipoVoto"]).agg({"votos":"sum"}).sort_values(["Cargo","votos"], ascending=False)
```

```python
# Quedamos solamente con los votos "positivos".
df1 = df[df.Agrupacion.notnull()]
```

```python
df1.shape
```

Totales de votos por Agrupación y cargo

```python
 df1.groupby(["Agrupacion", "Cargo"]).agg({"votos":"sum"}).sort_values("votos", ascending=False)
```

## Votos a Diputados PROVINCIALES por Agrupación

```python
df1[df1.Cargo == "DIPUTADOS PROVINCIALES"].groupby(["Agrupacion"]).agg({"votos":"sum"}).sort_values("votos", ascending=False)
```

```python
# Votos a Diputados Provinciales por Agrupación
# transforma a DF.
source = df1[df1.Cargo == "DIPUTADOS PROVINCIALES"].groupby(["Agrupacion"]).agg({"votos":"sum"}).sort_values("votos", ascending=False).reset_index()
source
```

```python
alt.Chart(source, width=480, height=250, title="Diputados PROVINCIALES | Chaco").mark_bar().encode(
    x=alt.X('votos:Q', title='Votos'), #  stack="normalize"),
     y=alt.Y('Agrupacion:N', sort='-x', title=''),   
    color=alt.Color('votos:N', title= 'Votos', sort='-x'),
    tooltip=['votos','Agrupacion'],
).interactive()
```

Composición de votos del cargo Dip. Provinciales por Agrupación y lista.

```python
source = df1[df1.Cargo == "DIPUTADOS PROVINCIALES"].groupby(["Agrupacion", "lista"]).agg({"votos":"sum"}).sort_values("votos",ascending=False).reset_index()
source
```

```python
# Composición de votos por lista de cada Agrupación
bars = alt.Chart(source, width=480, height=250, title="Votos por LISTA de cada Agrupación | Dip. PROVINCIALES | Chaco").mark_bar().encode(
    x=alt.X('votos:Q', stack='zero', title='Votos'),
    y=alt.Y('Agrupacion:N', sort='-x', title=''),
    color=alt.Color('votos:N', title= 'Votos', sort='-x'),
    tooltip=['votos', 'lista','Agrupacion'],
)

text = alt.Chart(source).mark_text(dx=-5, dy=3, color='green', ).encode(
    x=alt.X('sum(votos):Q', stack='zero'),
    y=alt.Y('Agrupacion:N', sort='-x'),
    detail='lista:N',
    #text=alt.Text('sum(votos):Q', format='0l'),
)

bars + text
```

Calculo Cant. Electores por MUNCIPIO

```python
df_ElecMun = df_circuitos.groupby(["Municipio"]).agg({"Electores":"sum"}).sort_values("Electores", ascending=False)
```

```python
df_ElecMun
```

```python
df_merged = pd.merge(df, df_circuitos, how="left", left_on="IdCircuito", right_on="CIRCUITO")
```

```python
df_merged.head()
```

```python
df_merged.shape
```

```python
df_Electores = pd.merge(df_merged, df_ElecMun, how="inner", left_on="Municipio", right_on="Municipio")
```

```python
df_Electores.head()
```

PORCENTAJES


```python
df_ElecMunDipNac = df_Electores[df_Electores.Cargo == "DIPUTADOS NACIONALES"].groupby(["Seccion","Municipio"]).agg({"votos":"sum","Electores_y":"min"}).reset_index() # .sort_values("votos", ascending=False)
```

```python
df_ElecMunDipNac
```

### Dip. NACIONALES: Porcentajes de participación por MUNICIPIO

```python
source = df_ElecMunDipNac
barras = alt.Chart(source, title="Porcentajes de Participación. Diputados NACIONALES | Chaco").mark_bar().encode(
    x=alt.X('porcentaje:Q', title='Porcentajes de participación'), #  stack="normalize"),
     y=alt.Y('Municipio:N', sort='-x', title=''),   
    color=alt.Color('color:N', title= 'Porcentajes', sort='-x'),
    #tooltip=['votos','Agrupacion'],
).transform_calculate(porcentaje='datum.votos / datum.Electores_y')


linea = alt.Chart(source).mark_rule(color='red').encode(
    x='mean(porcentaje):Q'
).transform_calculate(porcentaje='datum.votos / datum.Electores_y')

barras + linea
```

```python
df_ElecMunDipProv = df_Electores[df_Electores.Cargo == "DIPUTADOS PROVINCIALES"].groupby(["Seccion","Municipio"]).agg({"votos":"sum","Electores_y":"min"}).reset_index() # .sort_values("votos", ascending=False)
```

### Dip. PROVINCIALES: Porcentajes de participación por MUNICIPIO

```python
source = df_ElecMunDipProv
barras = alt.Chart(source, title="Porcentajes de Participación. Diputados PROVINCIAES | Chaco").mark_bar().encode(
    x=alt.X('porcentaje:Q', title='Porcentajes de participación'), #  stack="normalize"),
     y=alt.Y('Municipio:N', sort='-x', title=''),   
    color=alt.Color('color:N', title= 'Porcentajes', sort='-x'),
    #tooltip=['votos','Agrupacion'],
).transform_calculate(porcentaje='datum.votos / datum.Electores_y')


linea = alt.Chart(source).mark_rule(color='red').encode(
    x='mean(porcentaje):Q'
).transform_calculate(porcentaje='datum.votos / datum.Electores_y')

barras + linea
```
